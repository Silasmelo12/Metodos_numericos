```python
import numpy as np 
import matplotlib.pyplot as plt 

NP = 11
L = 1 
DX = L/(NP-1)
Alfa = 0.2
t = 0.5
DT = 0.01 
NT = int(t/DT)
V = Alfa * DT/(DX**2)
x = np.arange (0,L+DX, DX)

u_num = np.zeros ((NT,NP))

u_num[0,:]    = 100*np.sin(np.pi*x/L)
u_num[:, 0]   = 0 
u_num[:, -1]  = 0

for n in range (0, NT-1):
    for j in range (1, NP-1):
        u_num[n+1, j] = V * (u_num[n, j+1] - u_num[n+1, j] + u_num[n, j-1] + u_num[n+1, j])
        
plt.plot (u_num[10]);
```
